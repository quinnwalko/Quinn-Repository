```python
import pandas as pd
```

```python
file = "201801.csv"
file2 = "201802.csv"
file3 = "201803.csv"
file4 = "201804.csv"
file5 = "201805.csv"
file6 = "201806.csv"
```

```python
january_df = pd.read_csv(file)
february_df = pd.read_csv(file2)
march_df = pd.read_csv(file3)
april_df = pd.read_csv(file4)
may_df = pd.read_csv(file5)
june_df = pd.read_csv(file6)
```

```python
year2018_df = pd.concat([january_df, february_df, march_df, april_df, may_df, june_df])
year2018_df.head()
```

```python
year2018_df.to_csv("Output/year2018.csv", encoding="utf-8", index = False, header = True)
```
